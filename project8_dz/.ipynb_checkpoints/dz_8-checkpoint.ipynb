```python
from scipy import stats
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import StandardScaler
```

```python
z_scores = stats.zscore(df['Installs'])
z_scores
```

```python
threshold = 3
outliers = df['Installs'][z_scores > threshold]
df.loc[z_scores > threshold, 'Installs'] = df['Installs'].median()
```

```python
plt.hist(df['Installs'])
plt.xlabel('Installs')
plt.ylabel('Count')
plt.show()
```

```python
z_scores = stats.zscore(df['Reviews'])
threshold = 3
outliers = df['Installs'][z_scores > threshold]
df.loc[z_scores > threshold, 'Reviews'] = df['Installs'].median()
```

```python
#reviews_df = df.loc[df['Reviews'] > 1000000]
```

```python
plt.hist(df['Reviews'])
plt.xlabel('Reviews')
plt.ylabel('Count')
plt.show()
```

```python
df['Rating'] = winsorize(df['Rating'], limits=[0.05, 0.05])
```

```python
plt.hist(df['Rating'])
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
scaler = StandardScaler()
df[['Rating']] = scaler.fit_transform(df[['Rating']])
```

```python
df.head()
```

```python
plt.hist(df['Rating'])
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('googleplaystore_1.csv')
```

```python
le = LabelEncoder()
df['Type_Encoded'] = le.fit_transform(df['Type'])
df.head()
```

```python
df = pd.get_dummies(df, columns=['Content Rating'], prefix='ContentRating', drop_first=True)
```

```python
df.head()
```

```python
pivot_table = df.pivot_table(index='Category', columns='ContentRating_Teen', values='Rating', aggfunc='mean')
pivot_table
```

```python
pivot_table = df.pivot_table(index='Category', columns='ContentRating_Mature 17+', values='Rating', aggfunc='mean')
pivot_table
```
